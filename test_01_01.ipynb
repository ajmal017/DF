```python
import pandas as pd
df_symbols =pd.read_csv('/Users/d567533/DF/s-and-p-500-companies-financials/data/constituents.csv')
```

```python
df_symbols.head()
all_symbols = df_symbols['Symbol'].unique()
```

```python
# Import yfinance
import yfinance as yf  
 
# Get the data for the stock Apple by specifying the stock ticker, start date, and end date
data_map ={}
for symbol in all_symbols:
    
    data = yf.download(symbol)#,'2016-01-01','2018-01-01')
    data['Symbol'] = symbol
    if(len(data)>0):
        data_map[symbol] = data
# Plot the close prices
# import matplotlib.pyplot as plt
# data.Close.plot()
# plt.show()
```
